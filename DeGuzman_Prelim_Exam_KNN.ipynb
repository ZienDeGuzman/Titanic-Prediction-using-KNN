```python
#Import libraries
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('/content/train.csv')
dataset.head()
```

```python
print( len(dataset) )
```

```python
abs = dataset.dropna(subset = ['Age', 'Ticket', 'Fare','SibSp'])
abs.head()
```

```python
titanic = abs.drop(columns = ['PassengerId','Name','Cabin','Ticket','Embarked'])
```

```python
titanic.head()
```

```python
#assigning objects
titanic['Sex'].unique()
sex = {'male': 1, 'female': 0}
titanic['Sex'].replace(sex, inplace=True)
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
#Applying the KNN algorithm
X = titanic.drop(columns=['Survived'])
y = titanic['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
prediction = KNeighborsClassifier()
prediction.fit(X_train,y_train)
y_prediction=prediction.predict(X_test)
```

```python
#Display the accuracy
accuracy = accuracy_score(y_test, y_prediction)
accuracy
```
